```python
df = dataf.rename(columns= {'iyear':'Year','imonth':'Month', 'iday':'day','extended':'Extended','country_txt':'Country', 'provstate':'State','region_txt':'Region', 'city':'City','latitude':'Latitude','longitude':'Longitude','attacktype1_txt':'Attacktype', 'nkill':'Killed', 'nwound':'wounded', 'target1':'Target','summary':'Summary', 'gname':'Group','weaptype1_txt':'Weapontype' ,'motive':'Motive','targtype1_txt':'Targettype'})
```

```python
df =df[['Year','Month', 'day','Extended','Country', 'State','Region', 'City','Latitude','Longitude','Attacktype','Killed', 'wounded','Target','Summary', 'Group','Weapontype' ,'Motive','Targettype']]
```

```python
print('most attacked country: ', df['Country'].value_counts().idxmax())
print('most of the time type of attacks: ', df['Attacktype'].value_counts().idxmax())
print('Year in which most of the attack occured: ', df['Year'].value_counts().idxmax())
print('most attacked city:', df['City'].value_counts().index[1])
print('most attacked Region: ', df['Region'].value_counts().idxmax())
print('most attacked group: ', df['Group'].value_counts().idxmax())
print('month in which most of the attack occured: ', df['Month'].value_counts().idxmax())
```

```python
sns.heatmap(df.corr(), vmin= -1, vmax =1, cmap= sns.diverging_palette(20,220,as_cmap= True))
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(df['Year'],palette="rocket_r")
plt.title('Terriorist activity each year')
plt.show()
```

```python
TerrorRe = pd.crosstab(df.Year, df.Region)
TerrorRe.plot(kind = 'area',color = sns.color_palette('husl',12))
fig= plt.gcf()
fig.set_size_inches(20,10)
plt.show()
```

```python
df['wounded'] =df['wounded'].fillna(0).astype(int) #History of the Worst Terror Attacks in Heatmap using Plotly
df['Killed'] = df['Killed'].fillna(0).astype(int)
df['casualities'] = df['Killed'] + df['wounded']
```

```python
df1 = df.sort_values(by='casualities',ascending=False)[:40] #Values are sorted by the top 40 worst terror attacks as to keep the heatmap simple and easy to visualize
```

```python
heat=df1.pivot_table(index='Country',columns='Year',values='casualities')
heat.fillna(0,inplace=True)
```

```python
plt.subplots(figsize=(15,6)) #Top Countries affected by Terror Attacks
sns.barplot(df['Country'].value_counts()[:15].index,df['Country'].value_counts()[:15].values,palette='Blues_d')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

```python
plt.figure(figsize=(13,6)) #Methods of attack
sns.countplot(df['Attacktype'],data=df,order=df['Attacktype'].value_counts().index, palette='mako')
plt.xticks(rotation=90)
plt.xlabel('Method')
plt.title('Method of Attack')
plt.show()
```

```python
fig,axes = plt.subplots(figsize=(16,11),nrows=1,ncols=2) #Country and region
sns.barplot(x = df['Country'].value_counts()[:20].values, y = df['Country'].value_counts()[:20].index,ax=axes[0],palette = 'mako');
axes[0].set_title('Terrorist Attacks per Country')
sns.barplot(x=df['Region'].value_counts().values,y=df['Region'].value_counts().index, ax=axes[1])
axes[1].set_title('Terrorist Attacks per Region')
fig.tight_layout()
plt.show()
```

```python
plt.subplots(figsize=(11,10)) #Most active organisations
sns.barplot(y=df['Group'].value_counts()[1:12].index,x=df['Group'].value_counts()[1:12].values,palette='rocket_r')
plt.title('Most Active Terrorist Organizations')
plt.show()
```

```python
#Before and after the War on Terror

#After the 9/11 attacks in New York City, the US-led War on Terror was launched with the goal to distinguish terrorism primarily in the Middle East. 
#Here, it is shown as a turning point for activity patterns in each region.

df_after = df[df['Year']>=2001]
fig,ax = plt.subplots(figsize=(15,10),nrows=2,ncols=1)
ax[0] = pd.crosstab(df.Year,df.Region).plot(ax=ax[0])
ax[0].set_title('Change in Regions per Year')
ax[0].legend(loc='center left',bbox_to_anchor = (1,0.5))
ax[0].vlines(x=2001,ymin=0,ymax=7000,colors='red',linestyles='--')
pd.crosstab(df_after.Year,df_after.Region).plot.bar(stacked=True,ax=ax[1])
ax[1].set_title('After Declaration of War on Terror (2001-2017)')
ax[1].legend(loc='center left',bbox_to_anchor = (1,0.5))
plt.show()
```
